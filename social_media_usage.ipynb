```python
# Ques 18: Analyze whether users with high daily minutes spent also have higher posts per day.
```

```python
# Ans 18: 
correlation = df['Daily_Minutes_Spent'].corr(df['Posts_Per_Day'])

print("Correlation between daily minutes spent and posts per day:")
print(correlation)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Daily_Minutes_Spent', y='Posts_Per_Day', alpha = 0.6)
plt.title('Relationship between Daily Minutes Spent and Posts Per Day')
plt.xlabel('Daily Minutes Spent')
plt.ylabel('Posts Per Day')
plt.show()
```

```python
bins = [0, 30, 60, 90, 120, 150, 180, df['Daily_Minutes_Spent'].max()]
labels = ['0-30', '30-60', '60-90', '90-120', '120-150', '150-180', '180+']
df['Time_Bins'] = pd.cut(df['Daily_Minutes_Spent'], bins=bins, labels=labels)

avg_posts_per_bin = df.groupby('Time_Bins')['Posts_Per_Day'].mean()

plt.figure(figsize=(10, 6))
avg_posts_per_bin.plot(kind='line', marker='o', color='b')
plt.title('Average Posts Per Day vs. Daily Minutes Spent')
plt.xlabel('Daily Minutes Spent (Binned)')
plt.ylabel('Average Posts Per Day')
plt.grid(True)
plt.show()
```

```python
# Ques 19: Find and visualize the correlation between daily minutes spent, posts, likes, and follows.
```

```python
# Ans 19:
columns_of_interest = ['Daily_Minutes_Spent', 'Posts_Per_Day', 'Likes_Per_Day', 'Follows_Per_Day']
correlation_matrix = df[columns_of_interest].corr()

# Plot the heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of User Engagement Metrics')
plt.show()
```

```python
# Ques 20: Conduct a comparative analysis of posts and likes to see if higher posts correlate with higher likes across platforms.
```

```python
df.describe()
```

```python
# Ans 20:
def post_cat(value):
    if (value <= 4):
        return("0-4")
    elif (4 < value <= 8):
        return("5-8")
    elif (8 < value <= 12):
        return("9-12")
    elif (12 < value <= 16):
        return("13-16")
    elif (16 < value <= 20):
        return("17-20")

df["sorted_posts_pd"] = df["Posts_Per_Day"].apply(post_cat)
```

```python

category_order = ['0-4', '5-8', '9-12', '13-16', '17-20']

df['sorted_posts_pd'] = pd.Categorical(df['sorted_posts_pd'], categories=category_order, ordered=True)
```

```python
relation = df.groupby("sorted_posts_pd").agg("Likes_Per_Day").mean().sort_index()
```

```python
sns.lineplot(relation, marker="o")
```

```python
# Ques 21: Group users by platform and analyze trends in daily follows relative to posts and likes per day.
```

```python
# Ans 21:

platform_trends = df.groupby('App')[['Posts_Per_Day', 'Likes_Per_Day', 'Follows_Per_Day']].mean()
plt.figure(figsize=(12, 8))
sns.lineplot(platform_trends, markers=True)
plt.title("Trends in Daily Posts, Likes, and Follows by Platform")
plt.xlabel("Platform")
plt.ylabel("Average Daily Counts")
plt.grid(True)
```
